```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the CSV file into a DataFrame
df = pd.read_csv('Apple2009-2024.csv')
```

```python

# Clean up the column name by stripping extra spaces
df.columns = df.columns.str.strip()
```

```python
df.head(15)
```

```python
#  Remove dollar signs and commas from financial columns
df['Year Close Price'] = df['Year Close Price'].replace({r'\$': '', ',': ''}, regex=True).astype(float)
df['Revenue (millions)'] = df['Revenue (millions)'].replace({r'\$': '', ',': ''}, regex=True).astype(float)
df['EBITDA (millions)'] = df['EBITDA (millions)'].replace({r'\$': '', ',': ''}, regex=True).astype(float)
df['Net Income (millions)'] = df['Net Income (millions)'].replace({r'\$': '', ',': ''}, regex=True).astype(float)
df['PE ratio'] = df['PE ratio'].replace({',': ''}, regex=True).astype(float)
df['Gross Margin'] = df['Gross Margin'].replace({r'%': ''}, regex=True).astype(float)
```

# **visualization **                                       

```python


# Plot 1: Line Plot for Year-End Close Price
plt.figure(figsize=(10, 6))
plt.plot(df['year'], df['Year Close Price'], marker='o', label='Year Close Price', color='b')
plt.xlabel('Year')
plt.ylabel('Close Price ($)')
plt.title('Apple Year-End Close Price (2009-2024)')
plt.xticks(df['year'], rotation=45)
plt.grid(True)
plt.tight_layout()
plt.legend()
plt.show()
```

```python
# Plot 2: Bar Chart for Revenue Over the Years

plt.figure(figsize=(10, 6))
colors = plt.cm.viridis(np.linspace(0, 1, len(df)))
plt.bar(df['year'], df['Revenue (millions)'], color=colors)
plt.xlabel('Year')
plt.ylabel('Revenue (Millions)')
plt.title('Apple Revenue Over the Years')
plt.xticks(df['year'], rotation=45)
plt.tight_layout()
plt.show()
```

```python


# Plot 3: Pie Chart for Gross Margin (Last 5 Years)
recent_years = df.tail()
plt.figure(figsize=(7, 7))
plt.pie(recent_years['Gross Margin'], labels=recent_years['year'], autopct='%1.1f%%', startangle=140,shadow='True')
plt.title('Gross Margin Distribution Over the Last 5 Years')
plt.show()





```

```python

# Plot 4: Scatter Plot for EBITDA vs Net Income
plt.figure(figsize=(10, 6))
plt.scatter(df['EBITDA (millions)'], df['Net Income (millions)'], color='m', label='EBITDA vs Net Income')
plt.xlabel('EBITDA (Millions)')
plt.ylabel('Net Income (Millions)')
plt.title('EBITDA vs Net Income')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Plot 5: Histogram of PE Ratio Distribution
plt.figure(figsize=(10, 6))
plt.hist(df['PE ratio'], bins=10, color='g', edgecolor='black')
plt.xlabel('PE Ratio')
plt.ylabel('Frequency')
plt.title('Distribution of PE Ratio Over the Years')
plt.grid(True)
plt.tight_layout()
plt.show()
```
